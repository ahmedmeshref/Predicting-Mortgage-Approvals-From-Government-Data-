```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import preprocessing 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
%matplotlib inline 
```

```python
## calling the train and test datasets 
df = pd.read_csv("train_values.csv")
df_tar = pd.read_csv("test_values.csv")
```

```python
df.head(5)
df_tar.head(5)
df.shape
df_tar.shape
df_tar.head()
```

```python
##for the testing dataset
df.co_applicant.unique()
df["co_applicant"].replace(True, 1, inplace = True)
df["co_applicant"].replace(False, 0, inplace = True)
df["co_applicant"].unique()


##for the target dataset
df_tar.co_applicant.unique()
df_tar["co_applicant"].replace(True, 1, inplace = True)
df_tar["co_applicant"].replace(False, 0, inplace = True)
df_tar["co_applicant"].unique()

df = df.astype(float)
df_tar = df_tar.astype(float)
df.dtypes
df_tar.dtypes

df.drop("accepted", axis=1).apply(lambda x: x.corr(df.accepted))
```

```python
## drop the row id column
df.head()
df.drop("row_id", axis = 1, inplace = True)
df_tar.drop("row_id", axis = 1, inplace = True)
```

```python
df["county_code"].replace(-1, np.nan, inplace= True)
df.dropna(how = "any", inplace = True)
df.isnull().sum()
print(df.shape)

df_tar["county_code"].replace(-1, np.nan, inplace= True)
df_tar.dropna(how = "any", inplace = True)
print(df_tar.shape)
df_tar.head()
```

```python
# Transfomation for the test dataset
x = df.drop('accepted', axis = 1)
y = df["accepted"]
x = preprocessing.StandardScaler().fit(x).transform(x)

## Transfomation for the target dataset
X = preprocessing.StandardScaler().fit(df_tar).transform(df_tar)

## print corr between all columns and the target column (accepted) in the test dataset
df.drop("accepted", axis=1).apply(lambda x: x.corr(df.accepted))
df_tar.shape
```

```python
## SPLITING THE TESTING DATASET
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state= 4)
```

```python
#LR = LogisticRegression().fit(x_train, y_train)
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)  
mlp.fit(x_train, y_train.values.ravel())  
yhat = mlp.predict(x_test)
```

```python
print("accuracy of the model =", accuracy_score(y_test, yhat))
```
